```python
# Imports
import os
import subprocess
import collections
import math
import scipy
import essentia
import librosa
import librosa.display
import numpy as np
import essentia.standard as es
from scipy.io import wavfile
import pandas as pd
```

```python
import IPython
from pylab import plot, show, figure, imshow
%matplotlib inline
import matplotlib.pyplot as plt

# Visualization library
import seaborn as sns
sns.set(style="white", color_codes=True)
sns.set_context(rc={"font.family":'sans',"font.size":24,"axes.titlesize":24,"axes.labelsize":24})

import warnings 
warnings.filterwarnings("ignore")
```

```python
AUDIO_FILE = 'resources/Sassuolo 1-4 AC Milan Trionfo del Milan sul Sassuolo con doppietta di Suso Serie A (online-audio-converter.com)_cleaned.wav'
EXT = '.wav'
EXT_SIZE = len(EXT)
SAMPLE_RATE = 44100

WINDOW_SIZE = 0.02
HOP_SIZE = 0.01
```

```python
# File names data frames
PRAAT_FEATURES_DF_FN = 'resources/Sassuolo 1-4 AC Milan Trionfo del Milan sul Sassuolo con doppietta di Suso Serie A (online-audio-converter.com)_praat_features.csv'
```

```python
DTYPE_PRAAT = {'Time': float,
               'Silences': int,
               'Pitch': float,
               'Intensity': float,
               'Harmonicity': float,
               'Jitter': float,
               'Shimmer': float}
```

```python
# Load audio file
audio = es.MonoLoader(filename=AUDIO_FILE, sampleRate=SAMPLE_RATE)()
audio_normalized = audio / np.max(np.abs(audio))

sig_len = len(audio)
time_feature = np.linspace(0, len(audio)/float(SAMPLE_RATE), math.floor(len(audio) / (HOP_SIZE * SAMPLE_RATE)))
```

```python
# Load Praat features data frame
praat_features_df = pd.read_csv(PRAAT_FEATURES_DF_FN, dtype=DTYPE_PRAAT)
praat_features_df.describe()
```

```python
voiced_feature = np.abs(praat_features_df['Silences'] - 1)
```

```python
# Load Essentia classes to extract remaining features
spectrum = es.Spectrum()
energy = es.Energy()
mfcc = es.MFCC()
centroid_t = es.SpectralCentroidTime()
entropy = es.Entropy()
power_spectrum = es.PowerSpectrum()
flux = es.Flux()
roll_off = es.RollOff()
zero_crossing = es.ZeroCrossingRate()
onset = es.OnsetRate()
log_norm = es.UnaryOperator(type='log')
```

```python
# Create spectrum variable (non directly used feature)
spectrum_feature = []
```

```python
# Features variables
chroma_feature = []
mfccs_feature = []
melbands_feature = []
melbands_log_feature = []
short_term_energy_feature = []
short_term_entropy_feature = []
spectral_centroid_feature = []
spectral_spread_feature = []
spectral_entropy_feature = []
spectral_flux_feature = []
spectral_roll_off_feature = []
zero_crossing_rate_feature = []
onsets_feature = []
onset_rate_feature = []
```

```python
# Load and display Chroma
# cqt_lbs = np.abs(librosa.cqt(audio, SAMPLE_RATE))
# chroma_map_lbs = librosa.filters.cq_to_chroma(math.floor(sig_len / (HOP_SIZE * SAMPLE_RATE))) #cqt_lbs.shape[0])
# chroma_feature = chroma_map_lbs.dot(cqt_lbs)
# chroma_feature = librosa.util.normalize(chroma_feature, axis=0)

chroma_feature = librosa.feature.chroma_stft(y=audio, sr=SAMPLE_RATE, hop_length=math.floor(HOP_SIZE*SAMPLE_RATE))
chroma_feature = librosa.util.normalize(chroma_feature, axis=0)[:, :-1]

plt.figure(figsize=(30,8))
plt.plot()
librosa.display.specshow(chroma_feature, y_axis='chroma')
plt.colorbar()
```

```python
plt.figure(figsize=(20,8))

plt.plot()
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
ax = sns.boxplot(data=[(chroma_feature[i, :]) for i in range(chroma_feature.shape[0])])
```

```python
for frame in es.FrameGenerator(audio, frameSize=math.floor(WINDOW_SIZE*SAMPLE_RATE), hopSize=math.floor(HOP_SIZE*SAMPLE_RATE), lastFrameToEndOfFile=False, startFromZero=True):
    spec = spectrum(frame)
    spectrum_feature.append(spec)
    mfcc_bands, mfcc_coeffs = mfcc(spec)
    mfccs_feature.append(mfcc_coeffs)
    melbands_feature.append(mfcc_bands)
    melbands_log_feature.append(log_norm(mfcc_bands))
    short_term_energy_feature.append(energy(frame))
    short_term_entropy_feature.append(entropy(np.power(frame, 2)))
    spectral_centroid_feature.append(centroid_t(spec))
    spectral_spread_feature.append(np.std(spec))
    tmp_power_spectral_density = power_spectrum(spec)  # Spectrum must have even length
    tmp_power_spectral_density_normalized = tmp_power_spectral_density / np.sum(tmp_power_spectral_density)
    spectral_entropy_feature.append(entropy(tmp_power_spectral_density_normalized))
    spectral_flux_feature.append(flux(spec))
    spectral_roll_off_feature.append(roll_off(spec))
    zero_crossing_rate_feature.append(zero_crossing(frame))
    ons, onsr = onset(frame)
    onsets_feature.append(ons)
    onset_rate_feature.append(onsr)
```

```python
spectrum_feature = essentia.array(spectrum_feature)
mfccs_feature = essentia.array(mfccs_feature)
chroma_feature = essentia.array(chroma_feature).T
melbands_feature = essentia.array(melbands_feature)
melbands_log_feature = essentia.array(melbands_log_feature)
short_term_energy_feature = essentia.array(short_term_energy_feature)
short_term_entropy_feature = essentia.array(short_term_entropy_feature)
spectral_centroid_feature = essentia.array(spectral_centroid_feature)
spectral_spread_feature = essentia.array(spectral_spread_feature)
spectral_entropy_feature = essentia.array(spectral_entropy_feature)
spectral_flux_feature = essentia.array(spectral_flux_feature)
spectral_roll_off_feature = essentia.array(spectral_roll_off_feature)
zero_crossing_rate_feature = essentia.array(zero_crossing_rate_feature)
onsets_feature = essentia.array(onsets_feature)
onset_rate_feature = essentia.array(onset_rate_feature)
```
